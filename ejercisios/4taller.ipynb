```
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['image.cmap'] = 'gray'

def escala_de_grises(im):
    ruta=(im)
    im= Image.open(ruta)
    im2= im
    i=0
    while i< im2.size[0]:
        j=0
        while j < im.size[1]:
            r,g,b = im2.getpixel((i,j)) #obtener pixeles
            gris= (r+g+b)//3 #el promedio de esos pixeles
            pixel= tuple([gris,gris,gris])
            im2.putpixel((i,j), pixel) #poner en el pixel
            j+=1
        i+=1
    im2.save("grises.jpg") 
    matriz_im("grises.jpg")
    
    
def matriz_im(im3):
    ruta=(im3)
    im3= plt.imread(ruta)
    im3=Image.open(ruta)
    im3= im3.resize([600,512]) #tamaño image
    im3_array=np.asarray(im3) #matriz 3 dim
    plt.imshow(im3_array,vmin=0,vmax=1) #im3.show()
    im3_array=np.reshape(im3_array,(960, 960)).copy()
    f=3
    while f<=957:
        c=3
        if(f==0):
           im3_array[f, c] =mascara(borde4(f, im3_array))
        if(f==959):
            im3_array[f, c] =mascara(borde3(f, im3_array))
            
        while c<=957:
            im3_array[f+1, c+1] =mascara(im3_array[f:f+3,c:c+3])
            if(c==0):
                im3_array[f, c] =mascara(borde1(f, im3_array))
            if(c==959):
                im3_array[f, c] =mascara(borde2(f, im3_array))
            c=c+1
        f=f+1
    im3_array=np.reshape(im3_array,(512,600, 3))
    #print("")
    #print(im3_array[24:25, 0:3, 0:3])
    plt.imshow(im3_array,vmin=0,vmax=1)#255
    
    
def borde1(f,n):#IZQ
    n1=(n[f:f+3, -1:])
    n1=np.reshape(n1,[3])
    n2=(n[f:f+3 , 0:1])
    n2=np.reshape(n2,[3])
    n3=(n[f:f+3 , 1:2])
    n3=np.reshape(n3,[3])
    c=np.append(n1,n2)
    c=np.append(c,n3)
    c=np.reshape(c,(3,3))
    c=c.T
    return(c)
    
def borde2(f,n):#DERE
    n1=(n[f:f+3, -2:-1])
    n1=np.reshape(n1,[3])
    n2=(n[f:f+3 , -1:])
    n2=np.reshape(n2,[3])
    n3=(n[f:f+3 , 0:1])
    n3=np.reshape(n3,[3])
    c=np.append(n1,n2)
    c=np.append(c,n3)
    c=np.reshape(c,(3,3))
    c=c.T
    return(c)
    
def borde3(f,n):#INFE
    n1=(n[-2:-1, f:f+3])
    n1=np.reshape(n1,[3])
    n2=(n[-1: , f:f+3])
    n2=np.reshape(n2,[3])
    n3=(n[0:1 , f:f+3])
    n3=np.reshape(n3,[3])
    c=np.append(n1,n2)
    c=np.append(c,n3)
    c=np.reshape(c,(3,3))
    return(c)
    
def borde4(f,n):#SUP
    n1=(n[-1:, f:f+3])
    n1=np.reshape(n1,[3])
    n2=(n[0:1 , f:f+3])
    n2=np.reshape(n2,[3])
    n3=(n[1:2 , f:f+3])
    n3=np.reshape(n3,[3])
    c=np.append(n1,n2)
    c=np.append(c,n3)
    c=np.reshape(c,(3,3))
    return(c)   
    
def mascara(a):
    b=np.array([[1,1,1], 
                [1,1,1], 
                [1,1,1]])
    f=0
    while (f<3):
        c=0
        while(c<3):
            b[f,c]=a[f,c]*b[f,c]
            c=c+1
        f=f+1    
    return(int(np.sum(b)//9))
    
    
escala_de_grises("a.jpg")
```
