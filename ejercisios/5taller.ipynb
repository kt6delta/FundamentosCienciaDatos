```
from PIL import Image
import numpy as np
import random
import matplotlib.pyplot as plt  
      
img=Image.open("grises.jpg")
temp1=(img.size[0]//4)#f
temp2=(img.size[1]//4)#c
array=np.asarray(img).copy()
#array=array[:temp1,:temp2,:]
p=0
while p<15:
    r=True
    while (r):
        f= random.randint(0, 3)*temp1
        c= random.randint(0, 3)*temp2
        a=array[f: f+temp1 ,c: c+temp2 ,:].copy() 
        
        f1= random.randint(0, 3)*temp1
        c1= random.randint(0, 3)*temp2
        while (f==f1 or c==c1):
            f1= random.randint(0, 3)*temp1
            c1= random.randint(0, 3)*temp2
        b=array[f1: f1+temp1 , c1: c1+temp2, :].copy() 
        if(a.shape==b.shape):
            r=False
            array[f1: f1+temp1 , c1: c1+temp2, :]=a
            array[f: f+temp1 ,c: c+temp2 ,:]=b
        else:
            r=True
    p=p+1
    
plt.imshow(array[:3*temp1 ,:4*temp2 ,:],vmin=0,vmax=1)
```
