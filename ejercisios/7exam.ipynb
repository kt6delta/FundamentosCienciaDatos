```
#Nombre Mateo Bohorquez Rodiguez Cod:20162021299 Reto No:1
import numpy as np
#funcion para aliear los dos arreglos y compararlos sumando los valores en el vector de puntajes
def align(l,m,r):
    h=np.zeros(208)#expancion de la matriz 
    h[29:179]=l#adaptacion a la matris de 150 caracteres
    
    for x in range(r.size): #recorrimiento de la matriz ampliada
        for y in range(m.size): 
            if int(h[int(x+y)])==m[y]:#encontrando si hay incidencia para cada posicion
                r[x]=r[x]+1 #guardando puntajes en el vector de puntajes
    return r
##se encuentra el mayor en el vector de puntajes 
def findingI(bigest,r):
    aux=0
    for x in range(r.size): 
        if r[x]==bigest and aux==0 :
            aux=x
            print('mayor= ',bigest,' posicion= ',aux)
    return aux
## funcion para tranformar la matriz a cadenas 
def transform(l):
    k=[]
    for x in range(l.size):
        if(l[x]==0):
            k.append('-')
        if(l[x]==1):
            k.append('A')
        if(l[x]==2):
            k.append('B')
        if(l[x]==3):
            k.append('C')
        if(l[x]==4):
            k.append('D')       
    return k
## funcion para mostrar el alineamiento donde ocurre mayor incidencia 
def showInf(l,m,i):
    h=np.zeros(208)
    h[29:179]=l
    k=np.zeros(h.size)
    k[i:int(i)+30]=m
    print(transform(h))
    print(transform(k))
    #generando las dos matrices aleatorias con valores de numeros enteros del 1 al 4
l=np.array(np.random.randint(4, size=150)+1)
m=np.array(np.random.randint(4, size=30)+1)
#creando vector de puntajes
r=np.zeros(179)
#llenando el vector con los puntajes
r=align(l,m,r)

j=findingI(r.max(axis=0),r)
#mostrando el alineamiento de los dos vectores
showInf(l,m,j)
```
